```python
url = "./研究/対話破綻/projectnextnlp-chat-dialogue-corpus/json/init100/1409463385.log.json"
```

```python
dir_rests = "./研究/対話破綻/projectnextnlp-chat-dialogue-corpus/json/rest1046/*"
dir_inits = "./研究/対話破綻/projectnextnlp-chat-dialogue-corpus/json/init100/*"
path_file = "./研究/対話破綻/projectnextnlp-chat-dialogue-corpus/json/dailog.txt"
```

```python
import glob
```

```python
rest_files = glob.glob(dir_rests)
init_files = glob.glob(dir_inits)
```

```python
rest_files
```

```python
with open(url, "r") as f:
    data = json.load(f)
    d = pd.DataFrame(data)
```

```python
s = d["turns"]
for i in range(0,21):
    print(s[i]["utterance"])
```

```python
text_list = []
```

```python
for i in range(len(init_files)):
    with open(init_files[i], 'r') as datafile:
        data = json.load(datafile)
    d = pd.DataFrame(data)
    s = d["turns"]
    for i in range(0,21):
        text_list.append(s[i]['utterance'] + "\n")
    
```

```python
len(text_list)
```

```python
with open(path_file, 'w') as file:
    for i in range(len(text_list)):
        file.write(text_list[i])
```
