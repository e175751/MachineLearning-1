```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 2
H = 3
T = 20
```

```python
dh = np.ones((N,H))
np.random.seed(3)
Wh = np.random.randn(H,H)
#Wh = np.random.randn(H,H)*0.5
```

```python
norm_list = []
for t in range(T):
    dh = np.dot(dh,Wh.T)
    norm = np.sqrt(np.sum(dh**2)) / N
    norm_list.append(norm)
```

```python
norm_list
x = []
y = []
for i,v in enumerate(norm_list):
    x.append(i)
    y.append(v)
```

### 重みに0.5を✖︎場合

```python
plt.plot(x,y)
plt.show()
```

### 重みに何もかけない場合

```python
plt.plot(x,y)
plt.show()
```
